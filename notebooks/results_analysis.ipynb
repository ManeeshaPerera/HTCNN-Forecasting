```python
cluster_pc = pd.read_csv('../swis_ts_data/cluster_pc.csv').rename(columns = {'Unnamed: 0': 'pc'})
```

```python
ALL_SWIS_TS = [6000, 6005, 6006, 6007, 6008, 6009, 6010, 6011, 6014, 6015, 6017, 6018, 6019, 6020, 6021, 6023,
               6024,
               6025, 6026, 6027, 6028, 6029, 6030, 6035, 6036, 6038, 6041, 6050, 6053, 6054, 6055, 6056, 6058, 6059,
               6061, 6062, 6063, 6064, 6065,
               6066, 6069, 6073, 6084, 6090, 6101, 6102, 6103, 6104, 6105, 6107, 6108, 6109, 6110, 6111, 6112, 6123,
               6147, 6148, 6149, 6150, 6151, 6152, 6153, 6154, 6155, 6157, 6159, 6162, 6163, 6164, 6166, 6167, 6168,
               6170, 6171, 6173, 6208, 6210,
               6211, 6220, 6230, 6232, 6233, 6237, 6254, 6271, 6280, 6281, 6284, 6312, 6317, 6324, 6330, 6391, 6401,
               6426, 6430, 6432, 6509, 6528, 6530]
```

```python
cluster_num = cluster_pc.loc[cluster_pc['pc'].isin(ALL_SWIS_TS)]
```

```python
cluster_num.head()
```

```python
# len(ALL_SWIS_TS_NEW)
```

```python
clusters = {}

for pc in range(len(cluster_num)):
    pc_num = cluster_num.iloc[pc,]['pc']
    cluster_val = cluster_num.iloc[pc,]['cluster_num']
    if cluster_val in clusters:
        clusters[cluster_val].append(pc_num)
    else:
        clusters[cluster_val] = [pc_num]
```

```python
clusters
```

```python
for key in clusters:
    power = []
    day2 = []
    day3 = []
    day4 = []
    day5 = []
    day6 = []
    day7 = []

    pcs = clusters[key]
    if len(pcs)>1:
        for pc_val in pcs:
            data_power = pd.read_csv(f'../swis_ts_data/ts_data/{pc_val}.csv', index_col =0)
            power.append(data_power.iloc[:, 0:1])
            day2.append(data_power.iloc[:, 1:2])
            day3.append(data_power.iloc[:, 2:3])
            day4.append(data_power.iloc[:, 3:4])
            day5.append(data_power.iloc[:, 4:5])
            day6.append(data_power.iloc[:, 5:6])
            day7.append(data_power.iloc[:, 6:7])

        power_df = pd.DataFrame(pd.concat(power, axis=1).sum(axis=1)).rename(columns={0:'power'})
        day2_df = pd.DataFrame(pd.concat(day2, axis=1).sum(axis=1)).rename(columns={0:'day2'})
        day3_df = pd.DataFrame(pd.concat(day3, axis=1).sum(axis=1)).rename(columns={0:'day3'})
        day4_df = pd.DataFrame(pd.concat(day4, axis=1).sum(axis=1)).rename(columns={0:'day4'})
        day5_df = pd.DataFrame(pd.concat(day5, axis=1).sum(axis=1)).rename(columns={0:'day5'})
        day6_df = pd.DataFrame(pd.concat(day6, axis=1).sum(axis=1)).rename(columns={0:'day6'})
        day7_df = pd.DataFrame(pd.concat(day7, axis=1).sum(axis=1)).rename(columns={0:'day7'})


        save_data = pd.concat([power_df, day2_df, day3_df, day4_df, day5_df, day6_df, day7_df], axis=1).round(2)
        save_data.to_csv(f'../swis_ts_data/ts_data/cluster_{key}.csv')
```
